```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\tamil\Music\adult.data.csv" ,header=None, names=["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"])
```

```python
df.head()
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(10, 7))
sns.histplot(df["age"], bins=20, color="blue", kde=True)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution in Adult Dataset")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
education_counts = df["education"].value_counts()
sns.barplot(x=education_counts.index, y=education_counts.values, palette="viridis")
plt.xticks(rotation=45)
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.title("Education Level Distribution in Adult Dataset")
plt.show()
```
