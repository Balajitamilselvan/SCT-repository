```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\tamil\Music\data-4.csv")
df.head()
```

```python
df["2021 Deaths"] = df["2021 Deaths"].str.replace(",", "").astype(float)
df["2022 Deaths"] = df["2022 Deaths"].str.replace(",", "").astype(float)
```

```python
df_cleaned = df[~df["Category"].str.contains("share in Total", case=False)]
```

```python
sns.set_style("darkgrid")
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=["2021", "2022"], y=[df_cleaned["2021 Deaths"].sum(), df_cleaned["2022 Deaths"].sum()])
plt.title("Total Road Accident Fatalities (2021 vs 2022)")
plt.ylabel("Number of Deaths")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_cleaned_sorted = df_cleaned.sort_values("% Change in Deaths", ascending=False)
sns.barplot(x="% Change in Deaths", y="Category", data=df_cleaned_sorted, palette="coolwarm")
plt.title("% Change in Fatalities by Category (2021-2022)")
plt.xlabel("% Change in Deaths")
plt.ylabel("Category")
plt.show()
```
